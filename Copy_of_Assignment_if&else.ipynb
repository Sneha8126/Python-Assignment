```python
date_year = user_date[:4]
date_month = int(user_date[5:7])
print(date_year)
print(date_month)
if date_month <= 3:
    quarter = date_year + '-Q1'
elif date_month <= 6:
    quarter = date_year + '-Q2'
elif date_month <= 9:
    quarter = date_year + '-Q3'
else:
    quarter = date_year + '-Q4'

print(quarter)
```

```python
# Surveyed data

locality_data = {'Name' : ['Abhinav','Sohail','Dheeraj','Rashmi','Vaishali'],
                  'Age' : [42,29,35,28,32],
                  'gender' : ['Male', 'Female','Male','Female','Female'] ,
                  'Height' : [172,175,170,168,170],
                  'Weight' : [70,82,60,55,64]
                 }
```

```python
Height = locality_data['Height']
print(Height)
Weight = locality_data['Weight']
print(Weight)
```

```python
total_height = sum(Height)
print(total_height)
n1=len(Height)
print(n1)
```

```python
total_weight = sum(Weight)
print(total_weight)
n2=len(Weight)
print(n2)
```

```python
mean_height=total_height/n1
print(mean_height)
```

```python
mean_weight=total_weight/n1
print(mean_weight)
```

```python
# Median height calculation
sorted_data = sorted(Height)
if n1 % 2 == 1:
   median_height = sorted_data[n1 // 2]
else:
   median_height = (sorted_data[n1 // 2 - 1] + sorted_data[n1 // 2]) / 2
print(median_height)
```

```python
# Median weight calculation
sorted_data = sorted(Weight)
if n2 % 2 == 1:
   median_weight = sorted_data[n2 // 2]
else:
   median_weight = (sorted_data[n2 // 2 - 1] + sorted_data[n2 // 2]) / 2
print(median_weight)
```
