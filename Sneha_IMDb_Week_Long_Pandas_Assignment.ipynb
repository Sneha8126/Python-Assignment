### Question 10
Create a report showing Total Movies, Average Budget, Average Revenue, Average Profit, Maximum Profit, Minimum Profit.

```python
# Write your solution here
report = pd.DataFrame({
    "Total Movies": [df.shape[0]],
    "Average Budget": [df["budget"].mean()],
    "Average Revenue": [df["revenue"].mean()],
    "Average Profit": [df["profit"].mean()],
    "Maximum Profit": [df["profit"].max()],
    "Minimum Profit": [df["profit"].min()]
})
report
```

### Question 11
Create a final summary dashboard using Pandas.

```python
import matplotlib.pyplot as plt

labels = report.columns
values = report.iloc[0]

# Bar Graph
plt.figure(figsize=(10,5))
plt.bar(labels, values)

# LOGARITHMIC SCALE
plt.yscale('log')

plt.title("Movie Report Summary (Log Scale)")
plt.xlabel("Report Metrics")
plt.ylabel("Values (Log Scale)")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
