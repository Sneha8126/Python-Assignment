# Question 1

```python
temperatures = [34,40,29,33,42,37,39]
min_val = min(temperatures)
max_val = max(temperatures)
min_index = temperatures.index(min_val)
max_index = temperatures.index(max_val)
temperatures[min_index], temperatures[max_index] = temperatures[max_index], temperatures[min_index]
print(temperatures)
```

# Question 2

```python
prod_price_list = [400, 250, 800, 550, 600, 820, 720, 15000, 360, 250]
total = prod_price_list[0] + prod_price_list[1] + prod_price_list[2] + prod_price_list[3] + prod_price_list[4] + prod_price_list[5] + prod_price_list[6] + prod_price_list[7] + prod_price_list[8] + prod_price_list[9]
mean = total / len(prod_price_list)
print(f"Mean: {mean}")
sorted_prices = sorted(prod_price_list)
n = len(sorted_prices)
mid1 = sorted_prices[n // 2 - 1]
mid2 = sorted_prices[n // 2]
median_price = (mid1 + mid2) / 2
print("Median:", median_price)
```
