## Use the file company_sales_data.csv

```
import numpy as np
import pandas as pd

from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv('/content/Copy of company_sales_data.csv')
```

```
df.head()
```

```
import seaborn as sns
```

# <u>**Problem 1**</u>

### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```
# Your code here

months = df['month_number']
profit = df['total_profit']
plt.figure(figsize=(9,3))
plt.bar(months,profit,color=sns.color_palette('Set2'))
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Total Profit Per Month")
plt.xticks(months)
plt.show()

```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']

plt.figure(figsize=(10, 5))

bar_width = 0.35
index = np.arange(len(months))

plt.bar(index,facecream_sales,width=bar_width ,label = "Face cream Sales")
plt.bar(index+bar_width,facewash_sales,width = bar_width,label = "Face Wash Sales")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in Number")
plt.title("Face Cream and Face Wash Sales Data")
plt.xticks(index + bar_width / 2, months)

plt.legend()
plt.show()


```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here

months = df['month_number']
plt.figure(figsize=(10,6))

plt.plot(months,df['facecream'],label='Face cream Sales',marker='o')
plt.plot(months,df['facewash'],label='Face wash Sales',marker='o')
plt.plot(months,df['toothpaste'],label='Toothpaste Sales',marker='o')
plt.plot(months,df['bathingsoap'],label='Bathing soap Sales',marker='o')
plt.plot(months,df['shampoo'],label='Shampoo Sales',marker='o')
plt.plot(months,df['moisturizer'],label='Moisturizer Sales',marker='o')

plt.xlabel('Month Number')
plt.ylabel('Sales units per month')
plt.title('Sales data of all products')
plt.legend(loc='upper left')

plt.xticks(months)
plt.show()

```
