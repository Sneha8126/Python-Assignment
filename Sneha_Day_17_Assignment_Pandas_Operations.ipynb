# <b>Problem 5 : Print the company name which manufactured the most expensive car. Also print the price of the most expensive car

```python

df['price'] = pd.to_numeric(df['price'])
most_expensive_car = df.sort_values(by='price',ascending=False).iloc[0]
print("Company Name:", most_expensive_car['make'])
print("Price:", most_expensive_car['price'])
```

# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
def cal_updated_price(row):
   if(row['engine-location']=='front'):
     return row['price']
   else:
    return row['price']*2

df['Updated-Price'] = df.apply(cal_updated_price,axis=1)
df.head()
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers.

```python
door_map = {'four': 4, 'two': 2}
df['number_of_doors'] = df['num-of-doors'].map(door_map).fillna(0).astype(int)
df.head()
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
GermanCars = pd.DataFrame({'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]})
japaneseCars = pd.DataFrame({'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]})
row_concatenate = pd.concat([GermanCars,japaneseCars],axis=0)
print(row_concatenate)

```

```python
col_concatenate = pd.concat([GermanCars,japaneseCars],axis=1)
print(col_concatenate)
```
