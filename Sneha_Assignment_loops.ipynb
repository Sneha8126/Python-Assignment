**# Quesion 3**

```python
# Store the list of jumbled word in a variable
jumbled_word = ["hugdtck", "tsogbklesrawqcbjplu", "qkisrubskoaitbt", "nslaoetr", "oksinjioaendnjsdqwbh"]
```

```python
lengths = []
for word in jumbled_word:
    lengths.append(len(word))
print("Lengths of words:", lengths)
```

```python

alphabets = " ,a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z"
alphabets = alphabets.split(',')
char_list = []
for i in lengths:
    char_list.append(alphabets[i])
print("List of all characters:", char_list)
```

```python
# 6. Get the sorted list and print it
sorted_char_list = sorted(char_list)
print("Sorted character list:", sorted_char_list)
```

```python
complete_message = "".join(sorted_char_list)
print("Complete message:", complete_message)
```

**# Question 4**

```python
perfect_number_candidate = 28
divisor_sum = 0

for i in range(1, perfect_number_candidate):
    if perfect_number_candidate % i == 0:
        divisor_sum += i

if divisor_sum == perfect_number_candidate:
    print(f"{perfect_number_candidate} is a perfect number.")
else:
    print(f"{perfect_number_candidate} is not a perfect number.")
```

**# Question 5**

```python
movie_quote = "If you let my daughter go now, that'll be the end of it. I will not look for you, I will not pursue you. But if you don't, I will look for you, I will find you, and I will kill you."
```

```python
clean_quote = movie_quote.replace(",", "").replace(".", "")
words_list = clean_quote.split()
unique_words = set()
for word in words_list:
    if len(word) > 3:
        unique_words.add(word)
print("Unique words with length greater than 3:")
print(unique_words)
```
