## Methods

Methods are functions defined inside the body of a class. They are used to perform operations with the attributes of our objects. Methods are a key concept of the OOP paradigm. They are essential to dividing responsibilities in programming, especially in large applications.

You can basically think of methods as functions acting on an Object that take the Object itself into account through its *self* argument.

Let's go through an example of creating a Circle class:

```python
def area(r):
  return 3.14*r*r
```

```python
def suc(r):
  return 2*3.14*r
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def circumference(self):
        return 2 * 3.14 * self.radius
```

```python
my_circle = Circle(5)
```

```python
print(f"Radius: {my_circle.radius}")
print(f"Area: {my_circle.area()}")
print(f"Circumference: {my_circle.circumference()}")
```

# Polymorphism

##### We've learned that while functions can take in different arguments, methods belong to the objects they act on. In Python, *polymorphism* refers to the way in which different object classes can share the same method name, and those methods can be called from the same place even though a variety of different objects might be passed in. The best way to explain this is by example:

##### **Definition:** Polymorphism means "many forms," allowing different objects to respond to the same function call in their own way.
##### **Common Interface:** It allows different classes to be treated as objects of a common superclass, providing a consistent way to interact with them.
##### **Behavioral Flexibility:** A single method name can behave differently depending on the specific object it is acting upon.
##### **Analogy:** Think of a "Speak" command given to different animals; a dog will bark, and a cat will meow, but the command remains the same for both.

Screenshot 2026-06-27 225426.png

* #####  The "Person": Imagine a single object, like a "Person".

* ##### Multiple Behaviors: This same person can perform different actions or take on different roles depending on the situation.

* ##### The Concept: Just as one person can act as an employee in an office, a shopper in a store, or a student in a classroom, polymorphism allows a single piece of code (or object) to behave in different ways depending on how it is used.

```python
class Dog:
    def speak(self):
        # Specific behavior for Dog
        print("Dog barks")

class Cat:
    def speak(self):
        # Specific behavior for Cat
        print("Cat meows")

# A common function that uses polymorphism
def animal_sound(animal):
    # The same method 'speak' is called on different objects
    animal.speak()
```

```python
# Creating objects of different classes
dog = Dog()
cat = Cat()
```

```python
# Calling the function with different objects
# The function treats them as objects of a common interface
animal_sound(dog)
animal_sound(cat)
```

* # Hierarchial Inheritance

* # Hybrid Inheritance

```python
# Base class
class A:
    def method_A(self):
        print("Method from class A")

# Hierarchical inheritance: B and C inherit from A
class B(A):
    def method_B(self):
        print("Method from class B")

class C(A):
    def method_C(self):
        print("Method from class C")

# Multiple inheritance: D inherits from B and C
class D(B, C):
    def method_D(self):
        print("Method from class D")
```

```python
obj = D()
```

```python
obj.method_A()
obj.method_B()
obj.method_C()
obj.method_D()
```

* # Encapsulation

Encapsulation means wrapping data (variables) and methods (functions) together into a single unit (a class). It also involves hiding the internal state of an object from the outside world by using "private" variables.

**Data Hiding:** It restricts direct access to an object's internal state by making variables private.

**Controlled Access:** Access to private data is typically managed through public methods, such as getters and setters, ensuring that the data is modified only in authorized ways.

**Security:** By preventing external code from directly altering internal variables, it protects the object's integrity and prevents unintended side effects.

**Modularity:** Encapsulation makes code more modular, as the internal implementation details of a class can be changed without affecting the external code that uses it.

**Simplicity:** It reduces complexity by presenting a clear interface to the user while keeping the "how" of the data processing hidden.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable (hidden)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance
```

```python
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())
# print(account.__balance)    # This would cause an Error!
```

* # Abstraction

Abstraction is one of the four main principles of Object-Oriented Programming (OOP). It means hiding the implementation details and showing only the essential features of an object.



**For example:**

* When you drive a car, you use the steering wheel, brake, and accelerator.
* You don't need to know how the engine works internally.
* This is abstraction.

**Why use abstraction?**

* Hides unnecessary implementation details.
* Makes code easier to understand.
* Improves security by exposing only required functionality.
* Makes maintenance easier.

```python
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

# Child class
class Dog(Animal):
    def sound(self):
        return "Bark"

# Child class
class Cat(Animal):
    def sound(self):
        return "Meow"

```

```python
dog = Dog()
cat = Cat()
```

```python
print(dog.sound())
print(cat.sound())
```
